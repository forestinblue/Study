```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

text = open('C:/Users/kimjunseok/Downloads/practice_note/data/constitution.txt').read()
wordcloud = WordCloud().generate(text)

wordcloud.words_
```

## 2. masked

```python
#이미지 처리를 위한 패키지, 불용어 처리를 위한 패키지, 마스킹 이미지 불러오기
import numpy as np
from PIL import Image
from wordcloud import STOPWORDS

alice_mask = np.array(Image.open('C:/Users/kimjunseok/Downloads/practice_note/data/alice_mask.png'))
text = open('C:/Users/kimjunseok/Downloads/practice_note/data/alice.txt').read()

stopwords = set(STOPWORDS)
stopwords.add('said')

alice_mask
```

```python
# 마스킹 이미지 흑백 처리

plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap = plt.cm.gray , interpolation= 'bilinear')
plt.axis('off')
plt.show()
```

```python
#단어 빈도수 계산

wc = WordCloud(background_color='white',
               max_words = 2000, 
               mask = alice_mask,
               stopwords = stopwords)
wc = wc.generate(text)
wc.words_
```

```python
plt.figure(figsize=(12, 12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## 3. Image-colored wordcloud: 마스킹 이미지의 컬러를 사용한 wordcloud

```python
from wordcloud import ImageColorGenerator

alice_coloring = np.array(Image.open('C:/Users/kimjunseok/Downloads/practice_note/data/alice_color.png'))
image_colors = ImageColorGenerator(alice_coloring)

wc = WordCloud(background_color='white',
               max_words = 2000, 
               mask = alice_coloring,
               max_font_size = 40,
               random_state = 42,
               stopwords = stopwords)

wc.generate(text)
```

```python
plt.figure(figsize=(12, 12))
plt.imshow(wc.recolor(color_func=image_colors), interpolation='bilinear')
plt.axis('off')
plt.show()
```
