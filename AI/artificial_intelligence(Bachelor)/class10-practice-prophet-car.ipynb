## 학습

```python
model = Prophet()
model.fit(df)
```

## 향후 1년간 예측

```python
future = model.make_future_dataframe(periods=12, freq='MS')  #주기는 한달 MS
future
```

```python
forecast = model.predict(future)
model.plot(forecast)
plt.show()
```

```python
def plot_y_pred(y_test, y_pred):
    plt.figure(figsize=(8, 5))
    plt.plot(y_test, c='r')
    plt.plot(y_pred, c='b')
    plt.show()
    print("R2=", r2_score(y_test, y_pred).round(3))
    error = abs(y_test - y_pred)
    print("MAE=", error.mean().round(3))
    print("rmse=", np.sqrt((error**2).mean()).round(3))
    print("max=", max(error).round(3))
    
    #실제값과 예측치 비교
```

# 선형 또는 naive 모델

```python
def naive_model(x_seq):
    return np.array(x_seq).mean(axis=1)  #입력을 주어지면, 평균치로 return
```

```python
#하나 후 맞춤
def split_seq(sequence, n_steps):
    x, y = [], []
    for i in range(len(sequence)):
        if i > len(sequence) - n_steps - 1 : break
        x.append(sequence[i:i+n_steps])
        y.append(sequence[i+n_steps])
    return np.array(x), np.array(y)
```

```python
#먼 미래를 맞춤
def split_seq_2(sequence, n_steps_in, n_steps_out=1):
    x, y = [],[]
    for i in range(len(sequence)):
        end_ix = i + n_steps_in
        out_end_ix = end_ix + n_steps_out
        if out_end_ix > len(sequence):
            break
    seq_x, seq_y = sequence[i:end_ix], sequence[out_end_ix-1]
    x.append(seq_x)
    y.append(seq_y)
    return np.array(x), np.array(y)
```

```python
from sklearn.linear_model import LinearRegression
y = df.y.values  #y에 자동차 판매량이 나와있다
cut = np.int((len(y)*0.8))
train = y[:cut]
test = y[cut:]
y.shape, train.shape, test.shape, cut
```

```python
lr= LinearRegression()
for n_steps in range(1,20):
  x_train, y_train = split_seq(train, n_steps)
  x_test, y_test = split_seq(test, n_steps)
  lr.fit(x_train, y_train) 
  print(n_steps, lr.score(x_test, y_test).round(3))   #- 값은 시그마보다 편차가 크다는 뜻이다.
    #패턴을 찾기 어렵다
    #선형 모델에서는 패턴을 찾기 어렵다
    
    #선형모델에서는 과거 여러개를 보고 분석하기 어렵다 
```

```python
n_steps_in = 1
n_steps_out = 1
x_train, y_train = split_seq_2(train, n_steps_in, n_steps_out)
x_test, y_test = split_seq_2(test, n_steps_in, n_steps_out)
# lr.fit(x_train, y_train) 
y_pred = naive_model(x_test)
plot_y_pred(y_test, y_pred)

#한 달 전 것으로 다음달 비교

#이게 그나마 선형 모델중에 잘 나온 것 - 그래도 성능이 낮다
```

```python
n_steps_in = 9
n_steps_out = 1
x_train, y_train = split_seq_2(train, n_steps_in, n_steps_out)
x_test, y_test = split_seq_2(test, n_steps_in, n_steps_out)
# lr.fit(x_train, y_train) 
y_pred = naive_model(x_test)
plot_y_pred(y_test, y_pred)

#총 9달 것으로 다음달 비교 -> 성능이 매우 좋지 않다

#두루뭉실하게 평균적으로 예측해준다.
#트랜드를 따라가기에는 어렵다
```
