```python
from scipy.cluster.hierarchy import dendrogram, linkage

plt.figure(figsize=(10, 5))
link_dist = linkage(power, metric='euclidean', method='ward')  
dendrogram(link_dist, labels=power.index)
plt.show()
```

클러스터링 (KMeans)
--

```python
from sklearn.cluster import KMeans
k = KMeans(n_clusters=3).fit(power)
print(k.n_clusters)
```

```python
power['클러스터'] = k.labels_
power.head(10)
```

```python
power.drop('클러스터', axis = 1, inplace=True) ; power.head()
```

```python
centers = k.cluster_centers_ ; centers
```

```python
my_markers=['*','^', 'o','^','.',',','1','2']
my_color =['r','c','g','b','g','k','r','y']

plt.figure(figsize=(10, 8))
plt.xlabel('서비스업')
plt.ylabel('제조업')
for n in range(power.shape[0]):
    label = k.labels_[n]
    plt.scatter(power['서비스업'][n], power['제조업'][n], c=my_color[label], marker=my_markers[label], s=100)
    plt.text(power['서비스업'][n]*1.03, power['제조업'][n]*0.98, power.index[n])
    
for i in range(k.n_clusters):
    plt.scatter(centers[i][0], centers[i][1], c = 'r', s= 50)
```

### DBSCAN

```python
from sklearn.datasets import make_moons #make_moons패키지 가져오다
from sklearn.cluster import DBSCAN   #DBSCAN: 밀도로 묶는 함수
X, y = make_moons(n_samples=300, noise=0.05, random_state=11) #300개 샘플을 섞고 #노이즈를 적당히 섞고 #랜덤값 초기값으로 11 아무렇게 주다
#X에는 좌표값이 들어가고, y에는 0과 1이 들어간다.
```

```python
X[:5], y[:5]  #랜덤한 공간상의 점들, 이 점들이 0과 1의 값을 갖는다.
```

```python
plt.scatter(X[:,0], X[:,1], c=y) #X[:,0] 위에 x array에서 모든 항들의 0번째 값들
plt.show() # X[:,1] 위에 x array에서 모든 항들의 1번째 값들
#이 X들은 위치를 가리키고, y값은 랜덤한 값을 가리킨다.
# color를 y값으로 하겠다. 한 색은 0 한색은 1
```

```python
kmeans = KMeans(n_clusters=2) # kmeans로 클러스터링을 해보다, 클러스터 개수를 2개로 하다
predict = kmeans.fit_predict(X) # kmeans.fit_predict를 x값 가지고 하고
plt.scatter(X[:,0], X[:,1], c=predict) #predict: 0과 1로 클러스터링을 나눈 것, 예측한 것을 나눈 결과를 보는 것이다.
#kmeans는 거리가 가까운 것을 0과 1로 예측한다, 유클리디언 거리로 나눈 것이다
```

```python
dbscan = DBSCAN(eps=0.15, min_samples=5, metric='euclidean') #eps: 최소 반경
# min_samples: 밀도를 측정하기 위해서 반경안에 10개의 샘플이 있으면 하나의 그룹으로 하겠다.
predict = dbscan.fit_predict(X) #이러한 조건으로 dbscan하다
plt.scatter(X[:,0], X[:,1], c=predict)
#DBSCAN: 밀도 기반 클러스터링 알고리즘이다
```
